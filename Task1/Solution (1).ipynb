```python
import pandas as pd
```

```python
file1 = open("/content/test.txt", "r")

arr = []
while True:
    line = file1.readline()
    if not line:
        break
    arr.append(line.strip())
file1.close
```

```python
for i in range(890):
  arr[i] = float(arr[i][len(arr[i]) - 3:])
```

```python
arr[0]
```

```python
dataset = []
i = 0

subject = 0
week = 0
day = 0
time = 0
while i < 720:
  while subject < 10:
    while week < 2:
      while day < 6:
        while time < 6:
          dataset.append([arr[i], subject, week, day, time, subject // 2])
          time += 1
          i += 1
        else:
          time = 0
        day += 1
      else:
        day = 0
      week += 1
    else:
      week = 0
    subject += 1
  else:
    subject = 0
```

```python
dict_subject = {
  0: "Квантовая механика",
  1: "Квантовая теория информации",
  2: "Квантовые вычисления",
  3: "Сложность квантовых алгоритмов",
  4: "Квантовые алгоритмы в логистике",
  5: "Квантовое машинное обучение",
  6: "Моделирование квантовых систем",
  7: "Квантовые алгоритмы в химии",
  8: "Физическая реализация квантовых компьютеров",
  9: "Моделирование квантовых алгоритмов"
}
```

```python
dict_week = {
    0: "Неделя 1",
    1: "Неделя 2"
}
```

```python
dict_day = {
    0: "пн",
    1: "вт",
    2: "ср",
    3: "чт",
    4: "пт",
    5: "сб"
}
```

```python
dict_time = {
    0: "9:00-10:00",
    1: "10:00-11:00",
    2: "11:00-12:00",
    3: "12:00-13:00",
    4: "13:00-14:00",
    5: "14:00-15:00"
}
```

```python
dict_teacher = {
    0: "Иванов",
    1: "Петров",
    2: "Сидоров",
    3: "Карпов",
    4: "Соколов"
}
```

```python
dataset2 = []
for i in range(720):
  if dataset[i][0] == 1.0:
    dataset2.append(dataset[i])

for i in range(len(dataset2)):
  dataset2[i][1] = dict_subject[dataset2[i][1]]
  dataset2[i][2] = dict_week[dataset2[i][2]]
  dataset2[i][3] = dict_day[dataset2[i][3]]
  dataset2[i][4] = dict_time[dataset2[i][4]]
  dataset2[i][5] = dict_teacher[dataset2[i][5]]
```

```python
df = pd.DataFrame(dataset2, columns = ['first', 'Subject', 'Week', 'Day', 'Hours', 'Name'])
df
```

```python
df2 = df.iloc[: , 1:]
df2
```

```python
columns_titles = ["Week", "Day", "Hours", "Subject", "Name"]
df3 = df2.reindex(columns=columns_titles)
df3
```

```python
df3.to_csv('Solution.csv', encoding='utf-8')
```

```python
df3 = df3.sort_values(by=['Week', 'Day'])
df3['Group']='Группа 1'
df3
```

```python
def get_key(val):
    for key, value in dict_teacher.items():
        if val == value:
            return key


df4 = df3.copy()
df4['Name'] = df['Name'].map(lambda x: get_key(x))
```
