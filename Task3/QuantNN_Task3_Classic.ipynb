```python
import numpy as np
import glob
from PIL import Image as im
from numpy import asarray
```

```python
dataset = []
file = '/content/drive/MyDrive/Задача 3_ Сегментация изображений/*.png'
for filename in glob.glob(file):
  img = im.open(filename)
  arr = np.asarray(img)
  dataset.append(arr)
```

```python
import cv2
```

```python
binary = []
for arr in dataset:
   # _, thresh = cv2.threshold(arr, 50, 255, cv2.THRESH_BINARY)
   # thresh = cv2.adaptiveThreshold(arr, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
   blur = cv2.GaussianBlur(arr, (5,5), 0)
   # _, thresh = cv2.threshold(blur, 100, 255, cv2.THRESH_BINARY)
   _, thresh = cv2.threshold(blur, 255, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
   binary.append(im.fromarray(thresh))
```

```python
from PIL import Image
import matplotlib.pyplot as plt
for img in binary:
  plt.imshow(img)
  plt.show()
```
